```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
numpy.random.seed(13)
import pandas as pd
```

```python
df = pd.read_csv('prima.csv')
```

```python
df.columns=[0,1,2,3,4,5,6,7,8]
```

```python
dataset = numpy.loadtxt("prima.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
X[0:5]
```

```python
Y[0:5]
```

Now we create the model with 3 layers.
1 is the input layer with 8 inputs.
2 is the hidden layer.
3 is the output layer.

```python
model = Sequential()
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=150,batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
